```python
# Dados estatísticos básicos: medidas de tendência central

dados = {"jogador": ["Marcelo", "Pedro", "Marcelo", "Adriano", "Mauro", "Pedro", "Marcelo"],
         "infracao": ["FALTA VIOLENTA", "RECLAMAÇÃO", "FALTA COMUM", "RECLAMAÇAO", "FALTA COMUM", "FALTA VIOLENTA", "RECLAMAÇÃO"],
         "punicao": [4, 1, 3, 2, 4, 4, 2],
         "juiz_A": [4, 1, 3, 2, 4, 4, 2],
         "juiz_B": [2, 1, 4, 1, 1, 5, 6]
         }

df = pd.DataFrame(dados)
print(df)
print("")
print("Média: ", df["punicao"].mean())
print("Mediana: ", df["punicao"].median())
print("Moda: ", df["punicao"].mode().values)
```

```python
print("Juiz A")
print("----------")
print("Amplitude: ", df["juiz_A"].max() - df["juiz_A"].min())
print("Variância: ", f"{df['juiz_A'].var():.3f}")
print("Desvio padrão: ", f"{df['juiz_A'].std():.3f}")

print("\nJuiz B")
print("----------")
print("Amplitude: ", df["juiz_B"].max() - df["juiz_B"].min())
print("Variância: ", f"{df['juiz_B'].var():.3f}")
print("Desvio padrão: ", f"{df['juiz_B'].std():.3f}")
```

```python
ja_amplitude = round(df["juiz_A"].max() - df["juiz_A"].min(), 3)
ja_variancia = round(df['juiz_A'].var(), 3)
ja_desvio = round(df['juiz_A'].std(), 3)
ja_media = round(df["juiz_A"].mean(), 3)
ja_mediana = round(df["juiz_A"].median(), 3)
ja_moda = df["juiz_A"].mode().values


jb_amplitude = round(df["juiz_B"].max() - df["juiz_B"].min(), 3)
jb_variancia = round(df['juiz_B'].var(), 3)
jb_desvio = round(df['juiz_B'].std(), 3)
jb_media = round(df["juiz_B"].mean(), 3)
jb_mediana = round(df["juiz_B"].median(), 3)
jb_moda = df["juiz_B"].mode().values

print("  Valores      |   Juiz A   |   Juiz B  ")
print("----------------------------------------")
print(f"{'Média':<15}|{ja_media:^12}|{jb_media:^12}")
print(f"{'Mediana':<15}|{ja_mediana:^12}|{jb_mediana:^12}")
print(f"{'Moda':<15}|{ja_moda[0]:^12}|{jb_moda[0]:^12}")
print(f"{'Amplitude':<15}|{ja_amplitude:^12}|{jb_amplitude:^12}")
print(f"{'Variância':<15}|{ja_variancia:^12}|{jb_variancia:^12}")
print(f"{'Desvio padrão':<15}|{ja_desvio:^12}|{jb_desvio:^12}")
```

```python
notas = pd.DataFrame({
    "A1": [9.8, 7.2, 8.0],
    "A2": [5.3, 4.0, 3.5],
    "A3": [5.5, 8.1, 7.2],
    "A4": [7.0, 7.5, 6.5]},
    index=["P1", "P2", "P3"])

print("Notas finais: ")
print("--------------------------------")
print(notas)

print("\nMédia de cada aluno")
print("--------------------------------")
print(round(notas.mean(),1))

print("\nMaior nota de cada aluno")
print("--------------------------------")
print(notas.max())

print("\nMenor nota de cada aluno")
print("--------------------------------")
print(notas.min())

print("\nMédia de cada prova")
print("--------------------------------")
print(notas.mean(axis=1))

print("\nMaior nota de cada prova")
print("--------------------------------")
print(notas.max(axis=1))

print("\nMenor nota de cada prova")
print("--------------------------------")
print(notas.min(axis=1))
```

```python
dados = {
    "nadador": ["Simonas Bills",
                "Benjamin Proud",
                "Antony Ervin",
                "Florent Manaudou",
                "Andriy Hovorov",
                "Nathan Adrian",
                "Bruno Fratus",
                "Brad Tandy"],
    "nacionalidade": ["Lituania",
                      "Grã-Bretanha",
                      "Estados Unidos",
                      "França",
                      "Ucrânia",
                      "Estados Unidos",
                      "Brasil",
                      "África do Sul"],
    "tempo": [22.08, 21.68, 21.40, 21.41, 21.74, 21.49, 21.79, 21.79]
 }

raias = list(range(1,9))
prova50m = pd.DataFrame(dados, index=raias)
prova50m.index.name = "raia"
prova50m = prova50m.sort_values(by="tempo")

print("* * Resultado inal Ordenado por Tempo * *")
print(prova50m)

resultado_por_raia = prova50m["tempo"].rank(method="min")
print("\n* * Posição de cada nadador (por raia): ")
print(resultado_por_raia)
```
