```python
import pandas as pd
from google.colab import files

upload = files.upload()

flags_csv = list(upload.keys())[0]
```

```python
flags = pd.read_csv(flags_csv)

# Seleção

# Selecione apenas as linhas dos países da Oceania

v = (flags['landmass']==6)
flags_oceania = flags[v]

display(flags_oceania)
```

```python
# Projeção

# Selecione apenas as linhas dos países da Oceania com área acima de 200 mil km2

w = v & (flags['area'] > 200)
flags_oceania_area = flags[w]

df = flags_oceania_area[["name", "colours", "language", "landmass", "area"]]
df = df.reset_index(drop=True)

display(df)


```
