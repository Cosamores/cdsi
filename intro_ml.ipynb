```python
import pandas as pd
from google.colab import files

upload = files.upload()
flags_csv = list(upload.keys())[0]
```

```python
flags = pd.read_csv(flags_csv)
flags
print(flags.columns)
```

```python
# ----------------------------------------- #
# (1) - Conversão dos atributos do Grupo 2
#       De: Categóricos não binários
#       Para: Categóricos binários
# ----------------------------------------- #

for c in flags.columns:
  if c in ['landmass', 'zone', 'language', 'religion', 'mainhue', 'topleft', 'botright']:
    dummies = pd.get_dummies(flags[c], prefix=c)
    flags = flags.join(dummies)
```

```python
# ----------------------------------------- #
# (2) Normalização dos atributos dos grupos 3 e 4
#     De: Números contínuos e discretos
#     Para: Números com valores na faixa entre 0 e 1
# ----------------------------------------- #

for c in flags.columns:
  if c in ['area', 'population', 'bars', 'stripes', 'colours', 'circles', 'crosses', 'saltires', 'quarters', 'sunstars']:
    c_max = max(flags[c])
    c_min = min(flags[c])
    flags[c] = (flags[c] - c_min) / (c_max - c_min)
```

```python
# ------------------------------------------ #
# (3) - Exclusão dos atributos indesejados
# ------------------------------------------ #

flags = flags.drop(columns=['name', 'landmass', 'zone', 'language', 'religion', 'mainhue', 'topleft', 'botright'])
```

```python
# ----------------------------------------- #
# (4) - Imprimir a configuração final de flags
# ----------------------------------------- #

print('Flags head:')
flags.head()
print(flags.columns)
```

```python
from sklearn.neighbors import KNeighborsClassifier

# ------------------------------------------------ #
# Configurar os parâmetros e variáveis auxiliares
# ------------------------------------------------ #

# Valor de k a ser usado no k-NN
k = 5

# Nomes das labels
labels = ['red', 'green', 'blue', 'gold', 'white', 'black', 'orange']

# Quantidade de labels
q = len(labels)

# Número de registros na base de treinamento (=194)
N = flags.shape[0]

# ------------------------------------------------ #
# Cria um classificador com o algoritmo k-NN
# e faz a estimativa do desempenho preditivo
# com método LOO (Leave-One-Out)
# ------------------------------------------------ #

# for j: laço que percorre cada rótulo de classe
for j in range(0,q):
  print('---------------------------------------')
  print('Processando o Rótulo ', labels[j])
  print('---------------------------------------')

  # Instancia uma matriz de confusão para o rótulo j
  mc = pd.DataFrame({'predito_nao': [0,0],
                     'predito_sim':[0,0]},
                     index=['real_nao', 'real_sim'])

# ------------------------------------------------ #
# Divide a base de treinamento verticalmente
# em duas partes: X (atributos preditivos) e
# Y (atributos classe)

  X = flags.drop(columns=labels)
  Y = flags[labels[j]]

  # for i: laço que realiza o LOO para o rótulo j

  for i in range(0,N):
  # cria conjunto de treinamento
    X_treino = X.drop([i])
    Y_treino = Y.drop([i])

  # cria conjunto de teste
    X_test = X.iloc[[i],:]
    Y_test = Y.iloc[[i]]

    # treinamento do modelo k-NN com dados de treino
    modelo = KNeighborsClassifier(n_neighbors=k)
    modelo.fit(X_treino, Y_treino)

    # predição do modelo com dados de teste
    pred = modelo.predict(X_test)

    # Atualiza a matriz de confusão
    if (Y_test.iloc[0]==0):
      if (pred[0]==0):
        mc.loc['real_nao','predito_nao'] += 1
      if (pred == 1):
        mc.loc['real_nao','predito_sim'] += 1
    else:
      if (pred==0):
        mc.loc['real_sim','predito_nao'] += 1
      if (pred == 1):
        mc.loc['real_sim','predito_sim'] += 1

# Fim do LOO para o rótulo j:
# imprimir a matriz de confusão e acurácia
  print('Matriz de Confusão:')
  print(mc)
  acuracia = (mc.loc['real_sim','predito_sim'] + mc.loc['real_nao','predito_nao']) / N
  print('Acurácia: ', acuracia)
  print('---------------------------------------')

```
