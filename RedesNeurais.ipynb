```python
import pickle
import numpy as np
from sklearn.metrics import accuracy_score, classification_report
from yellowbrick.classifier import ConfusionMatrix
from sklearn.neural_network import MLPClassifier
```

```python
caminho = '/content/drive/MyDrive/IFSP/CDSI/'
```

```python
import pickle

with open(caminho+'credit.pkl', 'rb') as f:
  X_credit_treinamento, y_credit_treinamento, X_credit_teste, y_credit_teste = pickle.load(f)
```

```python
X_credit_treinamento.shape, y_credit_treinamento.shape
```

```python
X_credit_teste.shape, y_credit_teste.shape
```

```python
from sklearn.neural_network import MLPClassifier
```

```python
rede_neural_credit = MLPClassifier(
    max_iter=1500,
    verbose=True,
    tol=0.0000100,
    solver='adam',
    activation='relu',
    hidden_layer_sizes=(20, 20))
rede_neural_credit.fit(X_credit_treinamento, y_credit_treinamento)
```

```python
previsao_cred = rede_neural_credit.predict(X_credit_teste)

previsao_cred
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix

cm = confusion_matrix(y_credit_teste, previsao_cred)
cm
```

```python
with open(caminho+'census.pkl', 'rb') as f:
  X_census_treinamento, y_census_treinamento, X_census_teste, y_census_teste = pickle.load(f)
```

```python
X_census_treinamento.shape, y_census_treinamento.shape
```

```python
X_census_teste.shape, y_census_teste.shape
```

```python
rede_neural_census = MLPClassifier(
    max_iter=1500,
    verbose=True,
    tol=0.0000100,
    solver='adam',
    activation='relu',
    hidden_layer_sizes=(60,1000,60),
    learning_rate='adaptive')
rede_neural_census.fit(X_census_treinamento, y_census_treinamento)
```

```python
previsao_census = rede_neural_census.predict(X_census_teste)

previsao_census
```

```python
accuracy_score(y_census_teste, previsao_census)
```

```python
cm_census = ConfusionMatrix(rede_neural_census)
cm_census.fit(X_census_treinamento, y_census_treinamento)
cm_census.score(X_census_teste, y_census_teste)
cm_census.show()
```

```python
print(classification_report(y_census_teste, previsao_census))
```
