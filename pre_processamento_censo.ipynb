# Tratamento de atributos categóricos

## LabelEnconder
Transforma em números inteiros as categorias de cada atributo categórico não binário (mais de duas categorias)

```python
X_census[:,1]
```

```python
label_encoder_teste = LabelEncoder()
teste = label_encoder_teste.fit_transform(X_census[:,1])
teste
```

```python
X_census[0]
```

```python
label_encoder_workclass = LabelEncoder()
label_encoder_education = LabelEncoder()
label_encoder_marital_status = LabelEncoder()
label_encoder_occupation = LabelEncoder()
label_encoder_relationship = LabelEncoder()
label_encoder_race = LabelEncoder()
label_encoder_sex = LabelEncoder()
label_encoder_country = LabelEncoder()
```

```python
X_census[:,1] = label_encoder_workclass.fit_transform(X_census[:,1])
X_census[:,3] = label_encoder_education.fit_transform(X_census[:,3])
X_census[:,5] = label_encoder_marital_status.fit_transform(X_census[:,5])
X_census[:,6] = label_encoder_occupation.fit_transform(X_census[:,6])
X_census[:,7] = label_encoder_relationship.fit_transform(X_census[:,7])
X_census[:,8] = label_encoder_race.fit_transform(X_census[:,8])
X_census[:,9] = label_encoder_sex.fit_transform(X_census[:,9])
X_census[:,13] = label_encoder_country.fit_transform(X_census[:,13])
```

## OneHotEncoder

Transforma em números binários (0 e 1) as categorias de cada atributo categórico não binário (mais de 2 categorias).

Cada categoria de um atributo se transforma em uma nova columa (atributo). Exemplo: atributo "workclass", com 9 categorias, se transforma em 0 novos atributos. Cada linha terá valor 1 na coluna referente à categoria original e 0 nas outras colunas.

```python
np.unique(base_censo['workclass'], return_counts=True)
```

```python
len(np.unique(base_censo['workclass']))
```

```python
len(np.unique(base_censo['occupation']))
```

```python
# As outras colunas serão apagadas

onehotencoder = ColumnTransformer(transformers=[('OneHot', OneHotEncoder(), [1,3,5,6,7,8,9,13])], remainder='passthrough')
X_census = onehotencoder.fit_transform(X_census).toarray()
X_census
```

## Escalonamento dos valores

De: Valores com escalas diferentes (ex: age 17-99, education-num 1-16, hour-per-week 1-99)

Para: Valores com médias próximas a 0 e desvios padrão próximos a 1

```python
scaler_census = StandardScaler()
X_census = scaler_census.fit_transform(X_census)
X_census
```

```python
X_census_treinamento, X_census_teste, y_census_treinamento, y_census_teste = train_test_split(X_census, y_census, test_size=0.15, random_state=0)
```
