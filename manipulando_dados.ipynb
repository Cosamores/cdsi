```python
from google.colab import files
import pandas as pd

upload = files.upload()

file_name = list(upload.keys())[0]
```

```python
dados = pd.read_csv(file_name)

dados
```

```python
media = round(dados.mean(), 2)
mediana = dados.median()
moda = dados.mode()
moda_col = dados.mode(axis=1)
dp = round(dados.std(), 3)

print(f"Média:\n{media}")
print(f"\nMediana:\n{mediana}")
print(f"\nModa linhas:\n{moda}")
print(f"\nModa colunas:\n{moda_col}")
print(f"\nDesvio padrão:\n{dp}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure()
sns.boxplot(data=dados[['custo1', 'custo2']])
plt.title('Boxplot de custos (Seguradora 1 and Seguradora 2)')
plt.ylabel('Custo')
plt.show()
```

```python

plt.figure()
plt.plot(dados['idade'], dados['custo1'], marker='o', label='Seguradora 1')
plt.plot(dados['idade'], dados['custo2'], marker='o', label='Seguradora 2')
plt.title('Evolução do custo por idade')
plt.xlabel('Idade')
plt.ylabel('Custo')
plt.legend()
plt.grid(True)
plt.show()
```
