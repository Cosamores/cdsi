```python
from google.colab import files

upload = files.upload()

flags_csv = list(upload.keys())[0]

```

```python
flags = pd.read_csv(flags_csv)
flags
```

```python
# Obtendo as propriedades bsicas de cada atributo

i= 0
for c in flags.columns:
  i += 1
  att = flags[c]                          # atributo
  att_dtype = att.dtype                   # dtype
  att_tam_dominio = att.unique().size     # tamanho do domnio
  att_tem_nulo = any(att.isnull())        # valores nulos

  if(att_tam_dominio < 8):
    print("("+str(i)+") atributo: ", c, "\t",
          "dtype: ", att_dtype, "\t",
          "nulos: ", att_tem_nulo, "\n",
          "domnio: ", att.unique(), "\n")
  else:
    if(att_dtype == "object"):
      print("("+str(i)+") atributo: ", c, "\t",
            "dtype: ", att_dtype, "\t",
            "nulos: ", att_tem_nulo, "\n",
            "domnio (primeiros elementos): ", att.unique()[0:8], "\n")
    else:
      print("("+str(i)+") atributo: ", c, "\t",
            "dtype: ", att_dtype, "\t",
            "nulos: ", att_tem_nulo, "\n",
            "domnio (primeiros elementos): ", att.unique()[0:8], "\n"
            "media: ", round(att.mean(), 2), "\n",
            "desvio padro: ", round(att.std(), 2), "\n",
            "min: ", att.min(), "\n",
            "max: ", att.max(), "\n",
            "mediana: ", att.median(), "\n",
            "moda: ", att.mode(), "\n"
            )
#
```

```python
df_cores = pd.DataFrame()
for c in flags.columns:
  if c in ["red", "green", "blue", "gold", "white", "black", "orange"]:
    df_cores[c] = flags[c].value_counts()

df_cores
```

```python
list_cores = ["red", "green", "blue", "gold", "whitesmoke", "black", "orange"]

df_cores.plot(kind="barh", subplots=True, color=list_cores, figsize=(8, 25), ylabel="" )
plt.show()
```
